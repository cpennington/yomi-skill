```python
from model import YomiModel
import yomi.historical_record
from render import YomiRender
from nash import Nash

data_name, hist_games = yomi.historical_record.games(autodata="same")
fit_dir = f"fits/{data_name}"
min_games = 0
model = YomiModel(
    hist_games,
    fit_dir,
    "char_skill_elo_skill_deficit.stan",
    ["mu", "char_skill", "elo_logit_scale", "log_lik", "win_hat"],
    min_games,
)
render = YomiRender(data_name, model, 1500, 2000)
```

```python
nash = Nash.from_render(render)
```
