```python
from bacon import load_replay_results, compute_elo
```

```python
replays = load_replay_results('/bacon-replays')
```

```python
results = []

for low_k in range(15, 35, 1):
    for high_k in range(110, 140, 1):
        for k_cutoff in range(75, 150, 5):
            elo = compute_elo(replays, low_k=low_k, high_k=high_k, k_cutoff=k_cutoff)

            elo["points_diff"] = elo.elo_before_1 - elo.elo_before_2
            elo["win_chance"] = 1 / (1 + (-elo.points_diff / 1135.77).rpow(10))
            squared_error = (
                (elo.win_chance - elo.win).pow(2)
                + ((1 - elo.win_chance) - (1 - elo.win)).pow(2)
            ).sum()

            results.append((squared_error, low_k, high_k, k_cutoff))
            if low_k == high_k:
                break
```

```python
min(results)
```
